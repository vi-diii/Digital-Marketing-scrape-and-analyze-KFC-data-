```python
import datetime  as dt

delta=dt.timedelta(hours=7)
kfc['Date']=[(dt.datetime.fromisoformat(i.replace("T",' ').replace("Z",''))+delta) for i in kfc.pop('Date')]
kfc['Written_day']=[i.strftime('%Y-%m-%d') for i in kfc['Date']]
kfc['Written_time']=[i.strftime('%H:%M:%S') for i in kfc['Date']]
kfc.pop('Date')
kfc
```

```python
kfc_hashtag=kfc[['text','like','love','haha','wow','sad','angry','care']]
```

```python
import re
message_split=[[x.split('#') for x in re.split(',| |\t|\n',str(i)) if '#' in x] for i in kfc_hashtag['text']]
message_split=[['#'+e for x in i for e in x if e!=''] for i in message_split]
kfc_hashtag['message_split']=message_split
dict_count = (dict.fromkeys([x for i in message_split for x in i]))
dict_count = { i:dict.fromkeys(['like','love','haha','wow','sad','angry','care'],0) for i in dict_count.keys()}
```

```python
dict_count
kfc_hashtag
react_count_type=['like','love','haha','wow','sad','angry','care']
for i in kfc_hashtag.index:
  for index in kfc_hashtag.loc[i,'message_split']:
    for type_react in react_count_type:
      dict_count[index][type_react]+=kfc_hashtag.loc[i,type_react]
```

```python
df_hashtag_react=pd.DataFrame.from_dict(dict_count,orient='index')
df_hashtag_react
```
