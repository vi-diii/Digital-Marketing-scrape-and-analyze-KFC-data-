```python
# Expand reactions columns into multiple columns
post_df_full['reactions'] = post_df_full['reactions'].apply(lambda x : dict(eval(x)) )
post_df_full_reactions = post_df_full['reactions'].apply(pd.Series )
```

```python
# Merge expanded columns into dataframe
post_df_full_with_reactions = pd.concat([post_df_full, post_df_full_reactions], axis=1).drop('reactions', axis=1)
```
