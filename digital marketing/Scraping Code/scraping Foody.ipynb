```python
#example 
sun = 'https://www.foody.vn/ho-chi-minh/kfc-ba-hom'
#make a function to avoid others used variables
def StaticBurgerCrawl(driver,url):
  print(url)
  driver.get(url)
  #sleep(3)
  Re_dict={}
  #To crawl a lot of page, sleep is a bad method
  #And WebDriverWait solve it well
  Re_dict['Tên']=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//h1[@class='dt-name']"))).get_attribute('innerText')
  #too lazy to read full html then write a shorted XPATH, copy xpath seems faster ....
  try:
    Re_dict['Đánh giá']=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[@data-toggle='dropdown-score-detail']"))).get_attribute('innerText')
    Re_dict['Lượt đánh giá']=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//*[@id='basic-info']/div/div[1]/section/div[3]/div/div[1]/div[1]"))).get_attribute('innerText')
    sub_elmnt=WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='story-score-info grid-x align-middle']")))
    for i in range(len(sub_elmnt)):
      Re_dict[driver.find_element(By.XPATH,f"//div[@id='dropdown-score-detail']/div[{2+i}]/div[3]").get_attribute('innerText').strip()]=driver.find_element(By.XPATH,f"//div[@id='dropdown-score-detail']/div[{2+i}]/div[1]").get_attribute('innerText').strip()
  except:
    pass
  

  try:
    Re_dict['Mở cửa']=driver.find_element(By.XPATH,"//div[@class='time']").get_attribute('innerText').strip()
    Re_dict['Vị trí chi tiết']=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//*[@id='basic-info']/div/section[2]/div[1]/div[2]"))).get_attribute('innerText')
    Re_dict['Giá cả']=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//*[@id='basic-info']/div/section[2]/div[1]/div[4]"))).get_attribute('innerText')
  except:
    pass
  driver.get(url+'/thong-tin')
  #sleep(3)
  try:
    sub_elmnt=WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='info-item']")))
    for i in sub_elmnt:
      temp_list=(i.get_attribute('innerText').split('\n'))
      Re_dict[temp_list[0]]=temp_list[1]
  except:
    pass
  return Re_dict
  
print(StaticBurgerCrawl(driver,sun))
```
